```python
import requests
from bs4 import BeautifulSoup
import pprint 
import json
import re  
import demjson
```

```python
page_indexs = range(0, 250, 25)
regex = re.compile(r'\\(?![/u"])')  
```

```python
def get_html(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36"
    }
    return requests.get(url, headers=headers).text
```

```python
def download_all_htmls():
    htmls = []
    for idx in page_indexs:
        url = f"https://movie.douban.com/top250?start={idx}&filter="
        print(url)
        htmls.append(get_html(url))
    return htmls
```

```python
htmls = download_all_htmls()
```

```python
tmp = get_html("https://movie.douban.com/top250")
print(tmp)
```

```python
def get_new_urls(html):
    
    soup = BeautifulSoup(html, 'html.parser')
    article_items = (
        soup.find("div", class_="article")
            .find("ol", class_="grid_view")
            .find_all("div", class_="item")
    )
    datas = []
    for article_item in article_items:
        tmp = article_item.find("div", class_="pic").find("a", href=True)
        datas.append(tmp['href'])
    return datas
```

```python
new_urls = []
for html in htmls:
    tmp_urls = get_new_urls(html)
    for url in tmp_urls:
        new_urls.append(url)
```

```python
print(new_urls)
```

```python
def get_data(html):
    soup = BeautifulSoup(html, 'html.parser')
    head_content = soup.find("head")
    body_content = soup.find("body")
    
    strs = head_content.find("script", {"type": "application/ld+json"})
    ss = str(strs)
    
    ss = ss[ss.find('<script type="application/ld+json">') + len('<script type="application/ld+json">'):ss.rfind('</scripts>') - len('</scripts>')]
    fixed = regex.sub(r"\\\\", ss)  
    print(ss)
    data = json.loads(ss)
    detail_directory = body_content.find("div", class_="related-info").find("div", class_="indent")
    if detail_directory.find("span", class_="all hidden"):
        data['detailed_info'] = detail_directory.find("span", class_="all hidden").get_text().replace(' ', '')
    else:
        data['detailed_info'] = detail_directory.find("span").get_text().replace(' ', '')
        
    return data
```

```python
data_set = []
for new_url in new_urls:
    print(new_url)
    t = get_data(get_html(new_url))
    print(t['url'])
    data_set.append(t)
```
